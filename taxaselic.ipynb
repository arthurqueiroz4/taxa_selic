```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv("bcdata.sgs.1178.csv", sep=";")
```

```python
df1.head()
```

```python
print(type(df1['valor']))
taxa_selic = df1['valor'].values
data = df1['data'].values
print(type(taxa_selic))
```

```python
datas = list(df1['data'])
taxa = list(df1['valor'])
datas_taxa = list(zip(datas,taxa))
```

```python
dados_prontos = []
num = 0
for y in datas_taxa:
    if int(y[0][6:]) >= 2020:
        if y[0][4] != num:
            dados_prontos.append(y)
            num = y[0][4]
```

```python
datas_ok = []
taxa_ok = []
for x in dados_prontos:
    datas_ok.append(x[0][3:])
    taxa_ok.append(x[1])
    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.figure(figsize=(30, 10))
plt.xlabel('Data (d/m/a)')
plt.ylabel('Taxa selic (%)')
plt.title('Taxa selic')
plt.plot(datas_ok,taxa_ok)
plt.axis()
plt.grid(True)
plt.savefig('taxaselic.pdf', format='pdf')
plt.show()
```
